# **Завдання 2. Аналіз повноти відповідей респондентів**

```python
# Отримуємо список усіх питань з schema
questions = schema['qname'].unique()

print(f"Загальна кількість питань у schema: {len(questions)}")
questions[:10]
```

```python
# Інтерсекція множин: питання + колонки датасету

schema_questions_set = set(questions)
dataset_columns_set = set(df.columns)

common_questions = schema_questions_set.intersection(dataset_columns_set)

print(f"Кількість питань, які є і в schema, і в датасеті: {len(common_questions)}")
```

```python
# Відбираємо тільки питання з інтерсекції
df_questions = df[list(common_questions)]

df_questions.shape
```

```python
# Фільтрація респондентів без пропущених значень
complete_respondents = df_questions.dropna()

complete_respondents.shape
```

```python
# Кількість респондентів, які відповіли на всі питання
num_complete_respondents = complete_respondents.shape[0]

print(num_complete_respondents)
```

Додатково було проаналізовано кількість респондентів,
які заповнили не менше 90% питань опитування.

```python
# Частка заповнених відповідей для кожного респондента
fill_ratio = df_questions.notna().mean(axis=1)

# Респонденти з >= 90% заповнених відповідей
respondents_90 = df_questions[fill_ratio >= 0.9]

respondents_90.shape[0]
```

**Висновок:** У результаті аналізу повноти відповідей було встановлено, що жоден респондент не надав відповіді на всі запитання опитування без пропущених значень.
Це пояснюється великою кількістю умовних та необов’язкових питань у Stack Overflow Developer Survey, а також різною релевантністю окремих блоків питань для різних груп респондентів.
Отриманий результат є типовим для масштабних опитувань і свідчить про реалістичність та складність структури даних.

# **Завдання 7. Географічний аналіз компенсації Python-розробників**

```python
# Фільтруємо респондентів, які програмують на Python
python_devs = df[
    df['LanguageHaveWorkedWith']
    .str.contains('Python', case=False, na=False)
]
#Обираємо потрібні колонки
python_devs_comp = python_devs[['Country', 'ConvertedCompYearly']]
#Прибираємо пропущені значення компенсації
python_devs_comp = python_devs_comp.dropna(subset=['ConvertedCompYearly'])
#Групування та агрегація
country_compensation = (
    python_devs_comp
    .groupby('Country')['ConvertedCompYearly']
    .agg(
        mean_compensation='mean',
        median_compensation='median'
    )
    .reset_index()
)
#Сортування для наочності
country_compensation = country_compensation.sort_values(
    by='median_compensation',
    ascending=False
)
country_compensation = country_compensation.reset_index(drop=True)
country_compensation.head(10)
```

**Висновок:** Аналіз показує, що медіанна компенсація Python-розробників суттєво відрізняється між країнами. Найвищі значення спостерігаються у країнах з розвиненими IT-ринками та високою вартістю життя, що свідчить про сильну залежність рівня доходу від географічного фактору.

# **Завдання 9. Аналіз популярності Python по віковим категоріям**

```python
# 1. Фільтр: хто використовує Python
df['UsesPython'] = df['LanguageHaveWorkedWith'].str.contains(
    'Python',
    case=False,
    na=False
)

# 2. Групування по вікових категоріях
age_python_stats = (
    df
    .groupby('Age')
    .agg(
        total_respondents=('ResponseId', 'count'),
        python_users=('UsesPython', 'sum')
    )
    .reset_index()
)

# 3. Розрахунок відсотка Python-розробників
age_python_stats['python_percentage'] = (
    age_python_stats['python_users']
    / age_python_stats['total_respondents']
) * 100

# 4. Залишаємо тільки потрібні колонки
age_python_stats = age_python_stats[
    ['Age', 'python_percentage']
].sort_values(by='python_percentage', ascending=False)

age_python_stats
```

**Висновок:** У результаті аналізу популярності Python за віковими категоріями було встановлено, що найвищий відсоток Python-розробників спостерігається у віковій групі 18–24 роки. Це свідчить про високу популярність Python серед молодих спеціалістів та початківців у програмуванні. Водночас у старших вікових групах Python також залишається затребуваною мовою, що підтверджує її універсальність і широке застосування.

# **Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників**

```python
# 1. Видаляємо пропущені значення компенсації
df_comp = df.dropna(subset=['ConvertedCompYearly'])

# 2. Обчислюємо 75-й перцентиль компенсації
comp_75 = df_comp['ConvertedCompYearly'].quantile(0.75)

print(f"75-й перцентиль річної компенсації: {comp_75:.2f}")

# 3. Фільтруємо високооплачуваних віддалених працівників
high_paid_remote = df_comp[
    (df_comp['ConvertedCompYearly'] >= comp_75) &
    (df_comp['RemoteWork'] == 'Remote')
]

# 4. Аналіз популярності індустрій
industry_stats = (
    high_paid_remote['Industry']
    .value_counts()
    .reset_index()
)

industry_stats.columns = ['Industry', 'Count']

# 5. Виводимо топ-10 індустрій
industry_stats.head(10)
```

**Висновок:** У результаті аналізу було встановлено, що серед високооплачуваних віддалених працівників домінують індустрії Software Development, Fintech та Healthcare. Також суттєву частку становлять Internet, Telecom or Information Services і Banking/Financial Services. Це свідчить про те, що саме технологічні та фінансові сектори найчастіше поєднують високий рівень компенсації з віддаленим форматом роботи.
